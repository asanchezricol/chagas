Code to load observations and organise them

```python
sheet = pd.read_csv('nature_separated.csv', sep=',', encoding='latin-1')
```

```python
print(sheet)
```

```python
latitudes = sheet['decimalLatitude']
longitudes = sheet['decimalLongitude']
years = sheet['year']
species = sheet['species']
countries = sheet['countryCode']
states = sheet['stateProvince']
```

```python
#filter by year
years_indices = []
a = len(years)
for i in range(a): 
    if years[i]>1970: 
        years_indices.append(i)
```

```python
#filter by country
tx_indices = [] #'US'
co_indices = [] # 'Colombia'
mx_indices = [] #'Mexico' 
for j in years_indices: 
    b = countries[j] 
    if b == 'US': 
        if states[j]== 'Texas': 
            tx_indices.append(j)
    if b == 'CO': 
        co_indices.append(j)
    if b == 'MX': 
        mx_indices.append(j)

```

```python
header = ['species', 'countryCode','stateProvince','year','latitudes','longitudes']

#texas all##########
tx_species = []
tx_countryCode = []
tx_stateProvince  = []
tx_year = []
tx_latitudes = []
tx_longitudes = []

for i in tx_indices: 
    tx_species.append(species[i])
    tx_countryCode.append(countries[i])
    tx_stateProvince.append(states[i])
    tx_year.append(years[i])
    tx_latitudes.append(latitudes[i])
    tx_longitudes.append(longitudes[i])
    
#create DataFrame
df_tx = pd.DataFrame({'species': tx_species,
                   'countryCode': tx_countryCode,
                     'stateProvince':tx_stateProvince,
                      'year':tx_year,
                      'latitudes':tx_latitudes,
                      'longitudes':tx_longitudes})

#co all##########
co_species = []
co_countryCode = []
co_stateProvince  = []
co_year = []
co_latitudes = []
co_longitudes = []

for i in co_indices: 
    co_species.append(species[i])
    co_countryCode.append(countries[i])
    co_stateProvince.append(states[i])
    co_year.append(years[i])
    co_latitudes.append(latitudes[i])
    co_longitudes.append(longitudes[i])
    
#create DataFrame
df_co = pd.DataFrame({'species': co_species,
                   'countryCode': co_countryCode,
                     'stateProvince':co_stateProvince,
                      'year':co_year,
                      'latitudes':co_latitudes,
                      'longitudes':co_longitudes})

#mexico all##########
mx_species = []
mx_countryCode = []
mx_stateProvince  = []
mx_year = []
mx_latitudes = []
mx_longitudes = []

for i in mx_indices: 
    mx_species.append(species[i])
    mx_countryCode.append(countries[i])
    mx_stateProvince.append(states[i])
    mx_year.append(years[i])
    mx_latitudes.append(latitudes[i])
    mx_longitudes.append(longitudes[i])
    
#create DataFrame
df_mx = pd.DataFrame({'species': mx_species,
                   'countryCode': mx_countryCode,
                     'stateProvince':mx_stateProvince,
                      'year':mx_year,
                      'latitudes':mx_latitudes,
                      'longitudes':mx_longitudes})


```

```python
#create cvs texas
writer = pd.ExcelWriter('texas_all_timed.xlsx')
# write dataframe to excel
df_tx.to_excel(writer)
# save the excel
writer.save()
```

```python
#create cvs mexico
writer = pd.ExcelWriter('mexico_all_timed.xlsx')
# write dataframe to excel
df_mx.to_excel(writer)
# save the excel
writer.save()
```
